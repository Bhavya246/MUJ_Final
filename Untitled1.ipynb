Average True Range (ATR)

```python
def average_true_range(df, period=14):
    high_low = df['High'] - df['Low']
    high_close = abs(df['High'] - df['Close'].shift())
    low_close = abs(df['Low'] - df['Close'].shift())

    true_range = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)

    atr = true_range.rolling(window=period).mean()

    return atr
```

```python
combined_data['ATR'] = average_true_range(combined_data)
```

**LSTM (Long Short-Term Memory) Networks**

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error
```

```python
def create_sequences(data, seq_length):
    X = []
    y = []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)
```

```python
y_train_lstm = y_train_lstm[:, 0]
y_test_lstm = y_test_lstm[:, 0]
```

```python
print("Shape of X_train_lstm:", X_train_lstm.shape)
print("Shape of X_test_lstm:", X_test_lstm.shape)
print("Shape of y_train_lstm:", y_train_lstm.shape)
print("Shape of y_test_lstm:", y_test_lstm.shape)
```

**Feature Engineering Enhancements**

Moving Averages

```python
combined_data['MA_20'] = combined_data['Close'].rolling(window=20).mean()
combined_data['MA_50'] = combined_data['Close'].rolling(window=50).mean()
```

Exponential Moving Averages (EMA)

```python
combined_data['EMA_20'] = combined_data['Close'].ewm(span=20, adjust=False).mean()
combined_data['EMA_50'] = combined_data['Close'].ewm(span=50, adjust=False).mean()
```

MACD (Moving Average Convergence Divergence)

```python
combined_data['EMA_12'] = combined_data['Close'].ewm(span=12, adjust=False).mean()
combined_data['EMA_26'] = combined_data['Close'].ewm(span=26, adjust=False).mean()
combined_data['MACD'] = combined_data['EMA_12'] - combined_data['EMA_26']
```

Bollinger Bands

```python
combined_data['BB_upper'] = combined_data['MA_20'] + 2 * combined_data['Close'].rolling(window=20).std()
combined_data['BB_lower'] = combined_data['MA_20'] - 2 * combined_data['Close'].rolling(window=20).std()
```

```python
combined_data['Next_Close'] = combined_data['Close'].shift(-1)
combined_data.dropna(inplace=True)
```

```python
def create_sequences(data, seq_length):
    X = []
    y = []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

seq_length = 10
X_lstm, y_lstm = create_sequences(combined_data[features].values, seq_length)
split_index = int(len(X_lstm) * 0.8)
X_train_lstm, X_test_lstm = X_lstm[:split_index], X_lstm[split_index:]
y_train_lstm, y_test_lstm = y_lstm[:split_index], y_lstm[split_index:]

lstm_model = Sequential()
lstm_model.add(LSTM(50, activation='relu', input_shape=(seq_length, len(features))))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mse')
lstm_model.fit(X_train_lstm, y_train_lstm[:, 0], epochs=50, batch_size=32, verbose=1)
lstm_predictions = lstm_model.predict(X_test_lstm)
lstm_mse = mean_squared_error(y_test_lstm[:, 0], lstm_predictions)
print("LSTM MSE:", lstm_mse)
```
